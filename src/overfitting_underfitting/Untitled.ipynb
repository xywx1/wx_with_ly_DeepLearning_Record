```python
import math
import torch
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, TensorDataset
```

```python
# 设置设备为 CUDA
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
true_w = torch.zeros(20, 1).to(device)
true_w[0:4,0] = torch.tensor([2.1, 3, -2.5, 1.7]).to(device)
batch_size = 100
```

```python
def generate_data(dim=20):
    X = torch.normal(0, 1,size=(1000, dim)).to(device)
    poly_features = torch.pow(X, torch.arange(dim).to(device))
    for i in range(dim):
        poly_features[:, i] /= math.gamma(i+1)

    labels = poly_features @ true_w + torch.normal(0,0.1,size=(1000,1)).to(device)
    return X, labels
```

```python
def evaluate_loss(net, data_iter, loss):
    all_loss = []
    for X, y in data_iter:
        X, y = X.to(device), y.to(device)
        out = net(X)
        l = loss(out, y) #
        all_loss.append(l.mean().item())
    return sum(all_loss) / len(all_loss)
    
```

```python
def train(train_features, train_labels, num_epoch=20, lr=0.5, expected_dim=20):
    loss = torch.nn.MSELoss()

    train_f = train_features[:, :expected_dim].to(device)
    net = torch.nn.Sequential(torch.nn.Linear(expected_dim, 1, bias=False)).to(device)
    optimizer = torch.optim.SGD(net.parameters(), lr=lr)
    train_ls = []
    
    train_iter = getDataIter(train_f, train_labels)
    for epoch in range(num_epoch):
        net.train()
        all_loss = 0
        for X, y in train_iter:
            X, y = X.to(device), y.to(device)
            optimizer.zero_grad()
            l = loss(net(X), y)
            l.backward()
            optimizer.step()
            all_loss += l.mean().item()
        train_ls.append(all_loss)
        print('epoch %d, train loss %f' % (epoch + 1, train_ls[epoch]))
    plt.figure(figsize=(10, 5))
    plt.plot(range(num_epoch), train_ls, label='Train Loss')
    plt.show()
```

```python
def getDataIter(X, y):
    data = TensorDataset(X, y)
    iter = DataLoader(data, batch_size, shuffle=True)
    return iter
```

```python
train(*generate_data(), expected_dim=4)
```
